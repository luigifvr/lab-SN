In Statistical Mechanics the one-dimensional (1D) Ising Model is exactly resoluble. The Hamiltonian for a system of N spins with nearest neighbors interaction is given by (in the following I will assume units such that $\mu_B=1$ and $k_B=1$):

$$
H = 
-J\sum_{i=1}^N s_i s_{i+1}
-\frac{h}{2}\sum_{i=1}^N (s_i + s_{i+1}) 
$$

where $\{ s_i \}_{i=1,N}$ are spin variables taking values in $\{-1;1\}$, $J>0$ (ferromagnetic model),and with the assumption $s_{N+1} = s_{1}$.

The partition function for $N$ spins, $Z$, can be exactly computed and it turns out to be:

$$
Z=\sum_{\{ s_i \}} \exp \left( -\beta H\right) = \lambda_1^N + \lambda_2^N
$$

where
$\lambda_{1,2} = 
e^{\beta J} 
\cosh (\beta h) \pm \left[ e^{2\beta J} \cosh^2 (\beta h) - 2\sinh (2\beta J) \right]^{1/2}$, 
and where $\beta=\frac{1}{k_b T} = \frac{1}{T}$.
Once the partition function has been computed, the Helmholtz free energy is readly obtained as $A = -k_B T \ln Z$. All the thermodynamic quantities of interest can be deduced from $A$ or, equivalently, from $Z$:

- the **Internal Energy** per spin (with $h=0$):

$$
\frac{U(N,T)}{N} = \frac{\langle H \rangle}{N}
= \frac{1}{N} \sum_{\{ s_i \}} H \frac{\exp \left( -\beta H\right)}{Z}
= -\frac{1}{N}\frac{\partial \ln Z}{\partial \beta}
\underset{h=0}= -J \tanh (\beta J) \Biggl( \frac{1 + \tanh^{N-2} (\beta J)}{1+\tanh^N (\beta J)} \Biggl)
$$

```python
import numpy as np
import matplotlib.pyplot as plt
import glob as glob

def int_energy(x, nspins):
    beta = 1/x
    J = 1.0
    Ns = nspins
    th = np.tanh(J/x)
    thN= th**Ns
    ch = 1/th
    return -J*( th + ch*thN )/( 1 + thN )

def heat_capacity(x, nspins):
    beta = 1/x
    J = 1.0
    Ns = nspins
    th = np.tanh(J/x)
    thN= th**Ns
    ch = 1/th
    return ((beta*J)**2)*(((1+thN+(Ns-1)*(th**2)+(Ns-1)*(ch**2)*thN)/(1+thN))-Ns*((th+ch*thN)/(1+thN))**2)

def susceptibility(x, nspins):
    beta = 1/x
    J = 1.0
    Ns = nspins
    th = np.tanh(J/x)
    thN= th**Ns
    ch = 1/th
    return beta*np.exp(2*beta*J)*(1-thN)/(1+thN)

def magnetization(x, nspins, h_0):
    h = h_0 #external field
    J = 1.0
    b = 1/x
    Ns = nspins
    l1 = np.exp(b*J)*np.cosh(b*h)+np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))
    l2 = np.exp(b*J)*np.cosh(b*h)-np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))
    Z = l1**Ns + l2**Ns
    M = (np.exp(b*J)*np.sinh(b*h)*((l1**(Ns-1))*(1+np.exp(b*J)*np.cosh(b*h)/np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))) 
        + (l2**(Ns-1))*(1-np.exp(b*J)*np.cosh(b*h)/np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J)))))/(Z)
    return M
```

```python
files_gibbs_ene = glob.glob("ISING_1D/outputs/gibbs/output*.ene.0")
files_gibbs_heat = glob.glob("ISING_1D/outputs/gibbs/output*.heat.0")
files_gibbs_chi = glob.glob("ISING_1D/outputs/gibbs/output*.chi.0")
files_gibbs_mag = glob.glob("ISING_1D/outputs/gibbs/output*.mag.0")
gibbs_ene = []
gibbs_heat = []
gibbs_chi = []
gibbs_mag = []
for file_e, file_h, file_c, file_m in zip(files_gibbs_ene, files_gibbs_heat, files_gibbs_chi, files_gibbs_mag):
    gibbs_ene.append(np.genfromtxt(file_e, skip_header=99, usecols=(2,3)))
    gibbs_heat.append(np.genfromtxt(file_h, skip_header=99, usecols=(2,3)))
    gibbs_chi.append(np.genfromtxt(file_c, skip_header=99, usecols=(2,3)))
    gibbs_mag.append(np.genfromtxt(file_m, skip_header=99, usecols=(2,3)))
gibbs_ene = np.array(gibbs_ene)
gibbs_heat = np.array(gibbs_heat)
gibbs_chi = np.array(gibbs_chi)
gibbs_mag = np.array(gibbs_mag)
```

```python
files_metro_ene = glob.glob("ISING_1D/outputs/metro/output*.ene.0")
files_metro_heat = glob.glob("ISING_1D/outputs/metro/output*.heat.0")
files_metro_chi = glob.glob("ISING_1D/outputs/metro/output*.chi.0")
files_metro_mag = glob.glob("ISING_1D/outputs/metro/output*.mag.0")
metro_ene = []
metro_heat = []
metro_chi = []
metro_mag = []
for file_e, file_h, file_c, file_m in zip(files_metro_ene, files_metro_heat, files_metro_chi, files_metro_mag):
    metro_ene.append(np.genfromtxt(file_e, skip_header=99, usecols=(2,3)))
    metro_heat.append(np.genfromtxt(file_h, skip_header=99, usecols=(2,3)))
    metro_chi.append(np.genfromtxt(file_c, skip_header=99, usecols=(2,3)))
    metro_mag.append(np.genfromtxt(file_m, skip_header=99, usecols=(2,3)))
metro_ene = np.array(metro_ene)
metro_heat = np.array(metro_heat)
metro_chi = np.array(metro_chi)
metro_mag = np.array(metro_mag)
```

```python
points=100
temps = np.arange(0.5,2.1,0.1)
T = np.linspace(0.2,3.0,num=points)
e = int_energy(T, 50)
fig, (ax1, ax2) = plt.subplots(2,1, figsize=(8,6), gridspec_kw={'height_ratios':[3,1]}, sharex=True)
ax1.plot(T, e, label='True curve')
ax1.errorbar(temps, gibbs_ene[:,0], yerr=gibbs_ene[:,1], label='Gibbs sampler')
ax1.errorbar(temps, metro_ene[:,0], yerr=metro_ene[:,1], label='Metropolis')
ax1.set_title('Ising 1D, internal energy')
ax1.set_ylabel('U/N')
ax1.legend()
ax1.grid(True)

ax2.plot(temps, np.ones(len(temps)))
ax2.errorbar(temps, gibbs_ene[:,0]/int_energy(temps, 50), yerr=np.abs(gibbs_ene[:,1]/int_energy(temps, 50)))
ax2.errorbar(temps, metro_ene[:,0]/int_energy(temps, 50), yerr=np.abs(metro_ene[:,1]/int_energy(temps, 50)))
ax2.set_ylabel("Ratio to true value")
ax2.grid(True)
ax2.set_xlabel('T')
```

- the **Heat Capacity** (with $h=0$):

$$
C(N,T) = \frac{\partial U(N,T)}{\partial T}
= \frac{\partial \beta}{\partial T}\frac{\partial U}{\partial \beta}
= -k_B\beta^2 \frac{\partial}{\partial \beta} \left[ \sum_{\{ s_i \}} H \frac{\exp \left( -\beta H\right)}{Z} \right] = \\
= -k_B\beta^2 \left[ -\sum_{\{ s_i \}} H^2 \frac{\exp \left( -\beta H\right)}{Z} + \sum_{\{ s_i \}} H \exp \left( -\beta H\right)\frac{\partial Z^{-1}}{\partial \beta} \right] = \\
= k_B\beta^2 \left\{ \sum_{\{ s_i \}} H^2 \frac{\exp \left( -\beta H\right)}{Z} + \left[ \sum_{\{ s_i \}} H \frac{\exp \left( -\beta H\right)}{Z}\right]\frac{1}{Z}\frac{\partial Z}{\partial \beta} \right\} =\\
= k_B\beta^2 \left\{ \sum_{\{ s_i \}} H^2 \frac{\exp \left( -\beta H\right)}{Z} - \left[ \sum_{\{ s_i \}} H \frac{\exp \left( -\beta H\right)}{Z}\right]^2 \right\} =
k_B\beta^2 \left(\langle H^2 \rangle -\langle H \rangle^2 \right)
$$

$$
\frac{C(N,T)}{N} = \frac{1}{N}\frac{\partial U(N,T)}{\partial T} =\\
\underset{h=0}= 
k_B (\beta J)^2 \left\{ 1+\tanh^N (\beta J)+(N-1)\tanh^2 (\beta J)+(N-1)\frac{\tanh^{N-2} (\beta J)}{1+\tanh^N (\beta J)}
- N \left[ \frac{\tanh (\beta J) + \tanh^{N-1} (\beta J)}{1+\tanh^N (\beta J)} \right]^2 \right\}
$$

```python
heat_t= heat_capacity(T, 50)
fig, (ax1, ax2) = plt.subplots(2,1, figsize=(8,6), gridspec_kw={'height_ratios':[3,1]}, sharex=True)
ax1.plot(T, heat_t, label="True curve")
ax1.errorbar(temps, gibbs_heat[:,0], yerr=gibbs_heat[:,1], label="Gibbs sampler")
ax1.errorbar(temps, metro_heat[:,0], yerr=metro_heat[:,1], label="Metropolis")
ax1.set_title('Ising 1D, Heat capacity')
ax1.set_ylabel('C')
ax1.grid(True)
ax1.legend()

ax2.plot(temps, np.ones(len(temps)))
ax2.errorbar(temps, gibbs_heat[:,0]/heat_capacity(temps, 50), yerr=np.abs(gibbs_heat[:,1]/heat_capacity(temps, 50)))
ax2.errorbar(temps, metro_heat[:,0]/heat_capacity(temps, 50), yerr=np.abs(metro_heat[:,1]/heat_capacity(temps, 50)))
ax2.set_xlabel('T')
ax2.set_ylabel("Ratio to true value")
ax2.grid(True)
```

- The **magnetization** (with $h \not= 0$):

Note that with $h=0$ the magnetization $M(N,T,h=0)=\langle \sum_{i=1,N}s_i \rangle =0$ because in 1D we cannot have a (ferromagnetic) phase transition in absence of an external magnetic field.

$$
M(N,T,h) = -\frac{\partial A}{\partial h} = k_B T \frac{\partial \ln Z}{\partial h} =
\frac{1}{\beta} \frac{1}{Z}\frac{\partial}{\partial h}\sum_{\{ s_i \}} \exp \left( -\beta H\right) = \sum_{\{ s_i \}} \left[ \sum_{i=1,N}s_i\right] \frac{\exp \left( -\beta H\right)}{Z} = \left\langle \sum_{i=1,N}s_i \right\rangle
$$

The exact formula from $M(N,T,h) = k_B T \dfrac{\partial \ln Z}{\partial h}$ is:
$$
\frac{M(N,T,h)}{N}=\frac{e^{\beta J}\; \sinh(\beta h)}{Z} \; \Bigl[\lambda_1^{N-1}\cdot \Bigl(1+\frac{e^{\beta J}\; \cosh(\beta h)}{\sqrt{e^{2\beta J} \cosh^2 (\beta h) - 2\sinh (2\beta J)}}\Bigr) + \lambda_2^{N-1}\cdot \Bigl(1-\frac{e^{\beta J}\; \cosh(\beta h)}{\sqrt{e^{2\beta J} \cosh^2 (\beta h) - 2\sinh (2\beta J)}}\Bigr)\Bigr]
$$

```python
M = magnetization(T, 50, 0.02)
fig, (ax1, ax2) = plt.subplots(2,1, figsize=(8,6), gridspec_kw={'height_ratios':[3,1]}, sharex=True)
ax1.plot(T, M, label="True curve")
ax1.errorbar(temps, gibbs_mag[:,0], yerr=gibbs_mag[:,1], label="Gibbs sampler")
ax1.errorbar(temps, metro_mag[:,0], yerr=metro_mag[:,1], label="Metropolis")
ax1.set_title('Ising 1D, magnetization M with h = 0.02')
ax1.set_ylabel('$M$')
ax1.grid(True)
ax1.legend()

ax2.plot(temps, np.ones(len(temps)))
ax2.errorbar(temps, gibbs_mag[:,0]/magnetization(temps, 50, 0.02), yerr=np.abs(gibbs_mag[:,1]/magnetization(temps, 50, 0.02)))
ax2.errorbar(temps, metro_mag[:,0]/magnetization(temps, 50, 0.02), yerr=np.abs(metro_mag[:,1]/magnetization(temps, 50, 0.02)))
ax2.set_ylabel("Ratio to true value")
ax2.grid(True)
ax2.set_xlabel('T')
```

- The **magnetic susceptibility** (with $h=0$):

$$
\chi(N,T) = \frac{\partial M}{\partial h} = 
\frac{\partial}{\partial h}\sum_{\{ s_i \}} \left[ \sum_{i=1,N}s_i\right] \frac{\exp \left( -\beta H\right)}{Z} =\\
= \beta\sum_{\{ s_i \}} \left[ \sum_{i=1,N}s_i\right]^2 \frac{\exp \left( -\beta H\right)}{Z}
-\sum_{\{ s_i \}} \left[ \sum_{i=1,N}s_i\right] \frac{\exp \left( -\beta H\right)}{Z}
\frac{1}{Z}\frac{\partial Z}{\partial h} = 
\beta \left[ \left\langle \left( \sum_{i=1,N}s_i \right)^2 \right\rangle - \left\langle \sum_{i=1,N}s_i \right\rangle^2 \right] = \\
\underset{h=0}= \beta \left[ \left\langle \left( \sum_{i=1,N}s_i \right)^2 \right\rangle \right] = \beta e^{2\beta J} \Biggl( \frac{1-\tanh^N (\beta J)}{1+\tanh^N (\beta J)} \Biggl)
$$

```python
X = susceptibility(T, 50)
fig, (ax1,ax2) = plt.subplots(2,1, figsize=(8,6), gridspec_kw={'height_ratios': [3,1]}, sharex=True)
ax1.plot(T, X, label="True curve")
ax1.errorbar(temps, gibbs_chi[:,0], yerr=gibbs_chi[:,1], label="Gibbs sampler")
ax1.errorbar(temps, metro_chi[:,0], yerr=metro_chi[:,1], label="Metropolis")
ax1.set_title('Ising 1D, Susceptibility')
ax1.set_ylabel('$\chi$')
ax1.grid(True)
ax1.legend()

ax2.plot(temps, np.ones(len(temps)))
ax2.errorbar(temps, gibbs_chi[:,0]/susceptibility(temps, 50), yerr=gibbs_chi[:,1]/susceptibility(temps, 50))
ax2.errorbar(temps, metro_chi[:,0]/susceptibility(temps, 50), yerr=metro_chi[:,1]/susceptibility(temps, 50))
ax2.set_xlabel('T')
ax2.set_ylabel("Ratio to true value")
ax2.grid(True)
```
